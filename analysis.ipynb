```python
import json
results = json.load(open(r"results-v4.json"))
```

```python
from tqdm.notebook import tqdm
top20 = {}
for sheet_name in tqdm(results):
    top20[sheet_name] = {}
    allres = {}
    for model_id in tqdm(results[sheet_name], leave=False):
        tp = model_id.split('-')
        model_name = tp[0]
        random_seed = tp[-1]
        parameters = tp[1:-1]
        
        if model_name not in allres:
            allres[model_name] = []
        
        scores = results[sheet_name][model_id]
        r2_scores = scores['R2']
        aligned_r2_scores = scores['aligned r2']
        
        allres[model_name].append((r2_scores, aligned_r2_scores, parameters, random_seed))
    
    for model_name in tqdm(allres, leave=False):
        r2_top20 = sorted(allres[model_name], key=lambda x: x[0], reverse=True)[:20]
        ar2_top20 = sorted(allres[model_name], key=lambda x: x[1], reverse=True)[:20]
        top20[sheet_name][model_name] = {"r2_top20": r2_top20, "aligned_r2_top20": ar2_top20}
json.dump(top20, open("top20-v4.json", 'w'))
```

```python
top20 = json.load(open('top20-v4.json'))
top5 = {}
for sheet_name in tqdm(top20):
    top5[sheet_name] = {}
    for model_name in tqdm(top20[sheet_name], leave=False):
        top5[sheet_name][model_name] = {key.replace('_top20', '_top5'): [[x[0], x[1]] for x in value[:5]] for key, value in top20[sheet_name][model_name].items()}

print(json.dumps(top5, indent=2, ensure_ascii=False))
```

```python
top20 = json.load(open('top20-v4.json'))
top1 = {}
for sheet_name in tqdm(top20):
    top1[sheet_name] = {}
    for model_name in tqdm(top20[sheet_name], leave=False):
        top1[sheet_name][model_name] = {key.replace('_top20', '_top1'): value[0][0] if 'aligned' not in key else value[0][1] for key, value in top20[sheet_name][model_name].items()}

print(json.dumps(top1, indent=2, ensure_ascii=False))
```
